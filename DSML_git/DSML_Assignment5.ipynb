```python
cc=pd.read_csv("/content/CC GENERAL - CC GENERAL (1).csv")
cc.head()
```

```python
cc.shape
```

```python
cc.info()
```

```python
#histogram
plt.hist(cc['BALANCE'], bins=30, color='blue', edgecolor='black')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.title('Histogram ')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(9, 8))
sns.boxplot(data=cc[['PURCHASES', 'CASH_ADVANCE']], width=0.5)
plt.title('Boxplot of Purchases and Cash Advance', fontsize=16)
plt.show()
```

```python
#countplot
sns.countplot(x='TENURE', data=cc, palette='coolwarm')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.title('Count of Tenure')
plt.show()
```

```python
#barplot
mean_balance_by_tenure = cc.groupby('TENURE')['BALANCE'].mean()
mean_balance_by_tenure_cc = pd.DataFrame(mean_balance_by_tenure).reset_index()
plt.figure(figsize=(10, 6))
plt.bar(mean_balance_by_tenure_cc['TENURE'], mean_balance_by_tenure_cc['BALANCE'], color='blue')
plt.xlabel('Tenure')
plt.ylabel('Mean Balance')
plt.title('Mean Balance by Tenure')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#heatmap
corr_matrix = cc.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='plasma', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Credit Card Data')
plt.show()
```

```python
#pie chart
tenure_counts = cc['TENURE'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(tenure_counts, labels=tenure_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Tenure')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#lineplot
plt.figure(figsize=(10, 6))
plt.plot(cc['BALANCE'], label='Balance', color='blue', marker='o', linestyle='-')
plt.plot(cc['PAYMENTS'], label='Payments', color='green', marker='s', linestyle='-')
plt.title('Balance and Payments Over Time')
plt.xlabel('Index')
plt.ylabel('Amount')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
