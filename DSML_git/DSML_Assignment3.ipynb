```python
# Apply appropriate ML algorithm on a dataset collected in a cosmetics shop showing details of
# customers to predict customer response for special offer.
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/Mall_Customers - Mall_Customers.csv")
```

```python
data.head()
```

```python
data = pd.get_dummies(data,columns = ["Genre"])
```

```python
data.head()
```

```python
x = data.iloc[:,[0,1,2,4,5]]
print(x)
```

```python
y = data.iloc[:,[3]]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "linear", random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn import metrics
```

```python
r2_square = metrics.r2_score(y_test,y_pred)
mse = metrics.mean_squared_error(y_test,y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
print("R2 score: {0}\nMSE: {1}\nMAE: {2}".format(r2_square,mse,mae))
```
